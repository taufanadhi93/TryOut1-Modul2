```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('2019.csv')
df
```

```python
# Finding Central of Data Generosity

genInfo  = df['Generosity'].agg(Min=np.min, Max=np.max, Mean=np.mean, Median=np.median)
genMin   = df['Generosity'].min()
genMax   = df['Generosity'].max()
genMean  = df['Generosity'].mean()
genMedian= df['Generosity'].median()
genMode  = df['Generosity'].mode()


# Finding Distribution of Generosity Data

genSTD   = df['Generosity'].std()
genVAR   = df['Generosity'].var()
genQ1    = np.quantile(df['Generosity'],0.25)
genQ3    = np.quantile(df['Generosity'],0.75)
genIQR   = genQ3-genQ1
genLower = genQ1-(1.5*genIQR)
genUpper = genQ3+(1.5*genIQR)
genRange = genMax - genMin

print('Standard Deviation: ',genSTD)
print('Standard Variation: ',genVAR)
print('Q1                : ',genQ1)
print('Q3                : ',genQ3)
print('IQR               : ',genIQR)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.histplot(df['Generosity'])
ax.set_title('Generosity Distribution')
ax.set_xlabel('Generosity')
ax.set_ylabel('Frequency')

ax.axvline(x=genMean, label='Mean', color='r')
ax.axvline(x=genMedian, label='Median', color='b')
ax.legend()

plt.show()
```

```python
# Finding Central of Perception of corruption

perInfo  = df['Perceptions of corruption'].agg(Min=np.min, Max=np.max, Mean=np.mean, Median=np.median)
perMin   = df['Perceptions of corruption'].min()
perMax   = df['Perceptions of corruption'].max()
perMean  = df['Perceptions of corruption'].mean()
perMedian= df['Perceptions of corruption'].median()
perMode  = df['Perceptions of corruption'].mode()

# Finding Distribution of Perceptions of corruption Data

perSTD   = df['Perceptions of corruption'].std()
perVAR   = df['Perceptions of corruption'].var()
perQ1    = np.quantile(df['Perceptions of corruption'],0.25)
perQ3    = np.quantile(df['Perceptions of corruption'],0.75)
perIQR   = genQ3-genQ1
perLower = perQ1-(1.5*perIQR)
perUpper = perQ3+(1.5*perIQR)
perRange = perMax - perMin

print('Standard Deviation: ',perSTD)
print('Standard Variation: ',perVAR)
print('Q1                : ',perQ1)
print('Q3                : ',perQ3)
print('IQR               : ',perIQR)
print('Lower             : ',perLower)
print('Upper             : ',perUpper)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df,x='Perceptions of corruption')
ax.set_xlabel('Perceptions of corruption Value')
plt.show()
```

```python
# Data Dispersion

fig, ax = plt.subplots(figsize=(10,8))
sns.histplot(df['Perceptions of corruption'])
ax.set_title('Perceptions of corruption Distribution')
ax.set_xlabel('Perceptions of corruption')
ax.set_ylabel('Frequency')

ax.axvline(x=perMean, label='Mean', color='r')
ax.axvline(x=perMedian, label='Median', color='b')
ax.legend()

plt.show()
```
